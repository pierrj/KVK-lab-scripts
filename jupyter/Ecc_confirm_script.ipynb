```python
with open('10.reordered.putativeeccs', newline = '') as eccloc:
    eccloc_reader = csv.reader(eccloc, delimiter = '\t')
    eccloc_list = list(eccloc_reader)
    
with open('TESTFAKES4.discordantreads.bed', newline = '') as discordant:
    discordant_reader = csv.reader(discordant, delimiter = '\t')
    discordant_list = list(discordant_reader)
```

```python
def confirmeccs_nolist(output):
    with open(output, 'w', newline = '') as confirmed:
        w = csv.writer(confirmed, delimiter = '\t')
        for ecc in eccloc_list:
            for i in range(0, len(discordant_list), 2):
                read1 = discordant_list[i]
                read2 = discordant_list[i+1]
                if read1[0]==ecc[0]==read2[0] and ecc[1] <= read1[1] <= ecc[2] and ecc[1] <= read1[2] <= ecc[2] and ecc[1] <= read2[1] <= ecc[2] and ecc[1] <= read2[2] <= ecc[2]:
                    w.writerow(ecc)
                    break
```

```python
def confirmeccs_samechromfirst(output):
    with open(output, 'w', newline = '') as confirmed:
        w = csv.writer(confirmed, delimiter = '\t')
        for ecc in eccloc_list:
            for i in range(0, len(discordant_list), 2):
                read1 = discordant_list[i]
                read2 = discordant_list[i+1]
                if read1[0]==ecc[0]==read2[0]:
                    if ecc[1] <= read1[1] <= ecc[2] and ecc[1] <= read1[2] <= ecc[2] and ecc[1] <= read2[1] <= ecc[2] and ecc[1] <= read2[2] <= ecc[2]:
                        w.writerow(ecc)
                        break
```

```python
def confirmeccs_writeatend(output):
    confirmedeccs = []                
    for ecc in eccloc_list:
            for i in range(0, len(discordant_list), 2):
                read1 = discordant_list[i]
                read2 = discordant_list[i+1]
                if read1[0]==ecc[0]==read2[0]:
                    if ecc[1] <= read1[1] <= ecc[2] and ecc[1] <= read1[2] <= ecc[2] and ecc[1] <= read2[1] <= ecc[2] and ecc[1] <= read2[2] <= ecc[2]:
                        confirmedeccs.append(ecc)
                        break
    with open(output, 'w', newline = '') as confirmed:
        w = csv.writer(confirmed, delimiter = '\t')
        w.writerows(confirmedeccs)
```
