```python
def opensqldb():
    conn = sqlite3.connect(r"/global/home/users/pierrj/testfiles/sql.db")
    c = conn.cursor()
    return conn, c
```

```python
def data_entry(test):
    c.execute('''Drop TABLE if exists server''') ## to restart
    c.execute('''Create TABLE if not exists server(chrom, start, end)''')
    for i in range(len(test)):
        c.execute("INSERT INTO server(chrom, start, end) VALUES(?,?,?)", (test[i]))
    conn.commit()
```

```python
def testprint():
    c.execute("SELECT * FROM server")
    for row1 in c:
        row2 = next(c)
        print(row1[1])
        print(row2[1])
        print('hellooooo')
```

```python
def confirmeccs(ecc):
    c.execute("SELECT * FROM server")
    for read1 in c:
        read2 = next(c)
        if read1[0]==ecc[0]==read2[0]:
            if ecc[1] <= read1[1] <= ecc[2] and ecc[1] <= read1[2] <= ecc[2] and ecc[1] <= read2[1] <= ecc[2] and ecc[1] <= read2[2] <= ecc[2]:
                return True
    return False 
```

```python
def confirmeccs_nodb(ecc):
    for i in range(0, len(discordant_list), 2):
        read1 = discordant_list[i]
        read2 = discordant_list[i+1]
        if read1[0]==ecc[0]==read2[0]:
            if ecc[1] <= read1[1] <= ecc[2] and ecc[1] <= read1[2] <= ecc[2] and ecc[1] <= read2[1] <= ecc[2] and ecc[1] <= read2[2] <= ecc[2]:
                return True
    return False
```
